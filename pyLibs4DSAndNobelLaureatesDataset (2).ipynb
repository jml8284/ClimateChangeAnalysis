## Future Temperature Prediction

Predicts global average temperature changes for 2020-2030 using historical trends.


```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Prepare data
years = np.array(global_temp_trend.index).reshape(-1, 1)
temps = np.array(global_temp_trend.values).reshape(-1, 1)

# Create and fit the model
model = LinearRegression()
model.fit(years, temps)

# Predict future temperature changes
future_years = np.array(range(2020, 2031)).reshape(-1, 1)
future_temps = model.predict(future_years)

# Plot the results
plt.plot(years, temps, label="Historical Data")
plt.plot(future_years, future_temps, label="Predicted Data", linestyle="--", color="red")
plt.title("Global Temperature Change Trend and Prediction")
plt.xlabel("Year")
plt.ylabel("Average Temperature Change (°C)")
plt.legend()
plt.grid(True)
plt.show()
```
